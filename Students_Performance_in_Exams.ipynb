```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
sns.set(style="darkgrid", palette="bright", font_scale=1.5)
```

### Load Dataset
(Use "Students Performance in Exams" Dataset (kaggle))

```
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Python_exam/Data/StudentsPerformance.csv')

df.head()
```

```
df.describe()
```

```
sns.pairplot(df[['math score', 'reading score', 'writing score']], height=4) # sns.pairplot_데이터의 상관관계를 출력.
# 수학점수, 읽기점수, 쓰기점수 간의 상관관게. 막대 그래프는 분포도를 나타낸 것이고, 나머지 그래프는 점들이 모여 넓게 퍼진 것이 아닌, 일직선에 가까우면 상관관계가 높다는 것을 의미.(읽기점수와 쓰기점수와의 상관관계가 높음을 알 수 있음.)
```

### Compute Average Score

```
def average_score(dt):
    return(dt['math score'] + dt['reading score'] + dt['writing score']) / 3

df['average score'] = df.apply(average_score, axis=1) # axis의 1번 축으로 계산. 세 과목의 점수 평균을 구해 데이터프레임의 average score에 저장.
# df.apply()_데이터프레임에 함수를 일괄 적용.

df.head()
```

### Lunch and exam score

```
# sns.catplot()_categorical plot.
sns.catplot(x='lunch', y='math score', hue='gender', kind='boxen', data=df, height=10, 
            palette=sns.color_palette(['red', 'blue']))
plt.title('math')

sns.catplot(x='lunch', y='reading score', hue='gender', kind='boxen', data=df, height=10, 
            palette=sns.color_palette(['red', 'blue']))
plt.title('reading')

sns.catplot(x='lunch', y='writing score', hue='gender', kind='boxen', data=df, height=10, 
            palette=sns.color_palette(['red', 'blue']))
plt.title('writing')
```

### Test preparation course and exams average score

```
sns.catplot(x='test preparation course', y='average score', hue='gender', kind='boxen', data=df, height=10, 
            palette=sns.color_palette(['red', 'blue']))
plt.title('average')
```

### Parental level of education and exams average score

```
sns.catplot(x='parental level of education', y='average score', kind='boxen', data=df, height=14)
plt.title('average')
plt.legend(loc='lower right')
```
