```
book.info()
```

```
m = folium.Map([35.92, 128.60], zoom_start=13, control_scale = True)
m
```

```
# convert to (n, 3) nd-array format for heatmap
b_cpy = book[book["Count"] > 1].copy()
max_count = float(b_cpy['Count'].max())

b = list(zip(b_cpy['Latitude'], b_cpy['Longitude'], b_cpy['Count']))
```

```
# plot heatmap
m.add_child(plugins.HeatMap(b, radius=15, max_val = max_count))
m
```
